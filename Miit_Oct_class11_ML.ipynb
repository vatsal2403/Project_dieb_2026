```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/Diabetes_1.csv")
df
```

```python
# step 2 check weather the data is clean or not
# check missing values
df.isnull().sum()
```

```python
# check duplicates
df.duplicated().sum()
```

```python
# check data types
df.dtypes
```

```python
# check descriptive statistics
df.describe()
```

```python
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].median())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].median())
df["SkinThickness"]=df["SkinThickness"].replace(0,df["SkinThickness"].median())
df["Insulin"]=df["Insulin"].replace(0,df["Insulin"].median())
df["BMI"]=df["BMI"].replace(0,df["BMI"].median())
df
```

```python
l=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for i in l:
  df[i]=df[i].replace(0,df[i].median())
df
```

```python
df.describe()
```

```python
# step 3 Visualize the data

plt.figure(figsize=(20,7))
sns.boxplot(data=df)
```

```python
# heatmap
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm')
```

```python
# in real world you have to explore alot to understand the data
# you have to make 50-100 plots for every problem
```

```python
# step 3 split the data into input and target variable
X=df.drop('Outcome',axis=1) # input variables
y=df['Outcome'] # target variable
```

```python
# step 4 Split the input and an output into train & test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.20,
                                                    random_state=1)
```

```python
# step 6 Intialize the model
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression() # empty model and i am intializing the model
```

```python
# step 7 train the model
clf.fit(X_train,y_train)
```

```python
prob_ary=clf.predict_proba(X_test)
prob_ary
```

```python
y_pred=clf.predict(X_test)
```

```python
df1=pd.DataFrame({"Actual": y_test,
              "Predicted": y_pred,
              'ND_prob': prob_ary[:,0],
              'D_prob': prob_ary[:,1]})
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
# When data is imbalanced

# scenario 1: when false positive is important to reduce:-Precision
# scenario 2: when false negative is important to reduce:-recall
# scenario 3: when both false positive and false negative are
# important to reduce:-F1 score
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier() # intialize the model
clf.fit(X_train,y_train) # train the model
y_pred_rf=clf.predict(X_test) # test the model
```

```python
print("acc_score",accuracy_score(y_test,y_pred_rf))
print("precision_score",precision_score(y_test,y_pred_rf))
print("recall_score",recall_score(y_test,y_pred_rf))
print("f1_score",f1_score(y_test,y_pred_rf))
#
```
